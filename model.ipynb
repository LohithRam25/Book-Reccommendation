```python
new_data=pd.read_csv("D:\\MACHINE LEARNING\\DATA SETS\\books_data\\kindle_data-v2.csv")
```

```python
new_data.dropna(inplace=True)
```

```python
new_data.isnull().sum(
  
)
```

```python
new_data.head()
```

```python
new_data["features"]=new_data["category_name"]+" "+new_data["author"]+" "+new_data["stars"].astype(str)+" "+new_data["category_id"].astype(str)
```

```python
new_data["features"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
```

```python
vectors=vectorizer.fit_transform(new_data["features"])
```

```python
print(vectors)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(n_neighbors=10,metric="cosine")
```

```python
model.fit(vectors)
```

```python
from fuzzywuzzy import fuzz,process
```

```python
new_data.iloc[3]
```

```python
inpu1="wuthering heights"
titles=new_data["title"].tolist()
best_match = process.extractOne(inpu1, titles, scorer=fuzz.partial_ratio)
```

```python
best_match
```

```python
index=titles.index(best_match[0])
```

```python
index
```

```python
searched=new_data.iloc[3]
searched["title"]
```

```python
distances, indices = model.kneighbors(vectors[index].reshape(1, -1))
```

```python
indices
```

```python
for i in indices.flatten():
  print(titles[i])
```
